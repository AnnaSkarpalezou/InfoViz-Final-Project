```python
neighbourhoods = list(df.groupby('neighbourhood').groups.keys())
neighbourhoodsB = list(dfB.groupby('neighbourhood').groups.keys())
```

```python
stats_by_nei = df.groupby('neighbourhood')
n_listings_pernei = stats_by_nei[['host_id']].count()
n_listings_pernei
```

```python
stats_by_neiB = dfB.groupby('neighbourhood')
n_listings_perneiB = stats_by_neiB[['host_id']].count()
n_listings_perneiB
```

```python
room_type_by_nei = df.groupby(['neighbourhood','room_type'])[['id']].count()
room_type_by_nei
```

```python
room_type_by_neiB = dfB.groupby(['neighbourhood','room_type'])[['id']].count()
room_type_by_neiB
```

```python
new_df = pd.DataFrame({'Entire':room_type_by_nei['id'].iloc[::3].values, 'Private':room_type_by_nei['id'].iloc[1::3].values, 'Shared':room_type_by_nei['id'].iloc[2::3].values}, index = neighbourhoods)
new_df
    
```

```python
new_dfB = pd.DataFrame({'Entire':room_type_by_neiB['id'].iloc[::3].values, 'Private':room_type_by_neiB['id'].iloc[1::3].values, 'Shared':room_type_by_neiB['id'].iloc[2::3].values}, index = neighbourhoodsB)
new_dfB
```

```python
stats_by_nei = df.groupby('neighbourhood')
max_price = stats_by_nei[['price']].max()
min_price = stats_by_nei[['price']].min()
max_price.columns = ["maximum"]
min_price.columns = ["minimum"]
average_prices_by_nei.columns = ["Average"]
max_price


reviews_per_month = stats_by_nei[['reviews_per_month']].mean()
reviews_per_month.columns = ["Mean n of reviews per month"]
```

```python
stats_by_neiB = dfB.groupby('neighbourhood')
max_priceB = stats_by_neiB[['price']].max()
min_priceB = stats_by_neiB[['price']].min()
max_priceB.columns = ["maximum"]
min_priceB.columns = ["minimum"]
average_prices_by_neiB.columns = ["Average"]
max_priceB


reviews_per_monthB = stats_by_neiB[['reviews_per_month']].mean()
reviews_per_monthB.columns = ["reviews"]
```

```python
by_nei = pd.concat([average_prices_by_nei,max_price,min_price,new_df, reviews_per_month],axis=1)
by_nei
```

```python
by_nei.to_csv(r'/Users/annaskarpalezou/Desktop/InfoViz-Final-Project/iv-airbnb/public/data/tooltip_stats.csv', index = True)
```

```python
by_neiB = pd.concat([average_prices_by_neiB,max_priceB,min_priceB,new_dfB, reviews_per_monthB],axis=1)
by_neiB
```

```python
by_neiB.to_csv(r'/Users/annaskarpalezou/Desktop/InfoViz-Final-Project/iv-airbnb/public/data/tooltip_statsB.csv', index = True)
```
